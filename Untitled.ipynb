```python
from skimage import io
from skimage.metrics import structural_similarity
```

```python
test_files = glob('test_curated/*.jpg')
train_files = glob('train/*/*_image.jpg')

results = {}

for test_file in test_files:
    
    image = io.imread(test_file, as_gray=True)
    
    reference_ssim = structural_similarity(image, image, multichannel=True, gaussian_weights=True, sigma=1.5, use_sample_covariance=False, data_range=1.0)
    
    most_similar = None
    highest_ssim = -1
    
    folder = test_file.split('/')[1]
    folder = folder.split('_')[0]
    
    # find most similar one in train
    target_path = 'trainval/' + folder
    if os.path.exists(target_path):
        files = glob(target_path + '/*.jpg')
        for file in files:
            train_img = io.imread(file, as_gray=True)
            ssim = structural_similarity(image, train_img, multichannel=True, gaussian_weights=True, sigma=1.5, use_sample_covariance=False, data_range=1.0)
            
            if ssim > highest_ssim:
                highest_ssim = ssim
                most_similar = file
        
        results[test_file] = most_similar
    
    break
    
    
    
    
```
